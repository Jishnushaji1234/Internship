```python
#lets import all required librabries
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
#lets first connect to web driver
driver=webdriver.Chrome(r'C:/Users/jishn/Downloads/chromedriver_win32/chromedriver.exe')   
driver=webdriver.Chrome('chromedriver.exe')
url='http://www.flipkart.com/'
driver.get(url)
```

```python
search_product=driver.find_element_by_class_name("_3704LK")
search_product.send_keys("Sneakers")
```

```python
search_btn=driver.find_element_by_xpath('/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search_btn.click()
```

```python
#lets extract all web elements having job titles
title_tags=driver.find_elements_by_xpath('//div[@class="_2WkVRV"]')
titles=[]

for i in title_tags:
    titles.append(i.text)
```

```python
    
price_tags=driver.find_elements_by_xpath('//div[@class="_30jeq3"]')
price_names=[]

for i in price_tags:
    price_names.append(i.text)
```

```python
des_tags=driver.find_elements_by_xpath('//a[@class="IRpwTa"]')
des=[]

for i in des_tags:
    des.append(i.text)
```
