```python
#lets import all required librabries
import selenium
from selenium import webdriver
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
```

```python
#lets first connect to web driver
driver=webdriver.Chrome(r'C:/Users/jishn/Downloads/chromedriver_win32/chromedriver.exe')   
driver=webdriver.Chrome('chromedriver.exe')
url='http://www.myntra.com/shoes'
driver.get(url)
```

```python
Price_check=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[5]/ul/li[2]/label/div')
Price_check.click()
```

```python
Color_check=driver.find_element_by_xpath('/html/body/div[2]/div/div[1]/main/div[3]/div[1]/section/div/div[6]/ul/li[1]/label/div')
Color_check.click()                                         
```

```python
#lets extract all web elements having job titles
title_tags=driver.find_elements_by_xpath('//h3[@class="product-brand"]')
titles=[]

for i in title_tags:
    titles.append(i.text)
    
price_tags=driver.find_elements_by_xpath('//div[@class="product-price"]')
price_names=[]

for i in price_tags:
    price_names.append(i.text)
    
des_tags=driver.find_elements_by_xpath('//h4[@class="product-product"]')
des=[]

for i in des_tags:
    des.append(i.text)
```

```python
len(titles),len(price_names),len(des)
```

```python
Flipkart=pd.DataFrame()
Flipkart['Products']=titles
Flipkart['Description']=des
Flipkart['Price']=price_names
Flipkart
```
