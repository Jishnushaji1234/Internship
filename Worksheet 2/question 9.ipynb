```python
from bs4 import BeautifulSoup
import requests
page= requests.get('https://www.dineout.co.in/delhi-restaurants?search_str=best%20buffet%20special')
soup=BeautifulSoup(page.content)
titles =[]
for i in soup.find_all('div',class_="restnt-info cursor"):
    titles.append(i.text)
    
    
price =[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    price.append(i.text)
    
  
image =[]
for i in soup.find_all("img",class_="no-img"):
    image.append(i['data-src'])
    
    
print(len(titles),len(price),len(image))
     
#making a data frame
import pandas as pd
df= pd.DataFrame({'RESTAURANT AND LOCATION':titles,'PRICE':price,'IMAGES_URL':image})
df   
```
